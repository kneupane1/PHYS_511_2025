```python
# lets define a class called LoretzVector
class LorentzVector():
    def __init__(self,px,py,pz,energy):
            self.px = px
            self.py = py
            self.pz = pz
            self.energy = energy
    # Total Momentum P
    def P(self):
        return np.sqrt(self.px*self.px+self.py*self.py+self.pz*self.pz)
    # X,Y,Z,Energy-components of Momentum
    def Px(self): return self.px
    def Py(self): return self.py
    def Pz(self): return self.pz
    def Energy(self): return self.energy
    # Magnitude square of four vector
    def Mag2(self):
        return self.energy*self.energy-(self.px*self.px+self.py*self.py+self.pz*self.pz)
    
    # Magnitude of four vector
    def Mag(self):
        return np.sqrt(abs(self.Mag2())) 
    # It needs special tool to print 4-vector, it has 4 components
    def Print(self):
        print(self.px,self.py,self.pz,self.energy)
        return None 
    def Negate(self):
        self.px = -self.px
        self.py = -self.py
        self.pz = -self.pz
        self.Print()
        return None 
            
#Calculation of energy according to energy-mass relation, c=1
def Energy(px,py,pz,Mass):
    return np.sqrt(px*px+py*py+pz*pz+Mass*Mass)

#4-vector addition
def Add_4Vec(vec1,vec2):
    new_px = vec1.Px()+vec2.Px()
    new_py = vec1.Py()+vec2.Py()
    new_pz = vec1.Pz()+vec2.Pz()
    new_energy = vec1.Energy()+vec2.Energy()
    return LorentzVector(new_px,new_py,new_pz,new_energy)

#4-vector subtraction
def Sub_4Vec(vec1,vec2): # Vec1 - Vec2
    new_px = vec1.Px()-vec2.Px()
    new_py = vec1.Py()-vec2.Py()
    new_pz = vec1.Pz()-vec2.Pz()
    new_energy = vec1.Energy()-vec2.Energy()
    return LorentzVector(new_px,new_py,new_pz,new_energy)
```
